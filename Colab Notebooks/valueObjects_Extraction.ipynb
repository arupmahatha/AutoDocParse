```python
def remove_bounding_regions(json_data):
  """
  Recursively removes all not useful fields from a JSON object.
  """
  if isinstance(json_data, dict):
    if 'boundingRegions' in json_data:
      del json_data['boundingRegions']
    if 'polygon' in json_data:
      del json_data['polygon']
    if 'span' in json_data:
      del json_data['span']
    if 'columnIndex' in json_data:
      del json_data['columnIndex']
    if 'rowIndex' in json_data:
      del json_data['rowIndex']
    if 'confidence' in json_data:
      del json_data['confidence']
    if 'spans' in json_data:
      del json_data['spans']
    if 'kind' in json_data:
      del json_data['kind']
    for value in json_data.values():
      remove_bounding_regions(value)
  elif isinstance(json_data, list):
    for item in json_data:
      remove_bounding_regions(item)
```

```python
import json

def extract_value_objects(data):
    value_objects = []
    if isinstance(data, dict):
        for key, value in data.items():
            if key == 'valueObject':
                value_objects.append(value)
            else:
                value_objects.extend(extract_value_objects(value))
    elif isinstance(data, list):
        for item in data:
            value_objects.extend(extract_value_objects(item))
    return value_objects

def process_json(input_data):
    if isinstance(input_data, str):
        json_data = json.loads(input_data)
    elif isinstance(input_data, (bytes, bytearray)):
        json_data = json.loads(input_data.decode('utf-8'))
    else:
        raise TypeError('Input data must be a string, bytes, or bytearray.')

    # ... Rest of the function remains the same ...

    # Find the 'Items' key recursively
    items_data = find_items(json_data)

    if items_data:
        # Extract 'valueArray' from 'Items'
        value_array = items_data.get('valueArray', [])

        # Extract 'valueObject' from 'valueArray'
        value_objects = extract_value_objects(value_array)

        output_data = {
            'valueObjects': value_objects
        }

        return json.dumps(output_data, indent=4)
    else:
        return "No 'Items' found in the input JSON."

def find_items(data):
    # Recursively search for 'Items' key
    if isinstance(data, dict):
        if 'Items' in data:
            return data['Items']
        else:
            for key, value in data.items():
                result = find_items(value)
                if result:
                    return result
    elif isinstance(data, list):
        for item in data:
            result = find_items(item)
            if result:
                return result
    return None

if __name__ == "__main__":
    input_data = json.dumps(data)  # Replace this with your JSON input string or bytes

    processed_json = process_json(input_data)
    print(processed_json)
```
