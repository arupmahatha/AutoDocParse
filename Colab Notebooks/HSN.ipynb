```python
import re
import pickle
import pandas as pd
import json

def extract_hsn_codes(description):
    # Define the regular expression pattern to match "HS CODE" followed by any separator
    pattern = r'HS\s*CODE\s*[:\-]?\s*'
    # Find all occurrences of the pattern in the description
    matches = re.findall(pattern, description)

    # Extract the HS codes after each match
    hsn_codes = []
    for match in matches:
        # Find all HS codes after the match (separated by commas or spaces)
        hs_codes = re.findall(r'\d{4,10}', description.split(match, 1)[1])
        hsn_codes.extend(hs_codes)

    return hsn_codes

# Function to find numbers with length between 4 and 11 characters in a string
def find_numbers(text):
    numbers = re.findall(r'\b\d{4,11}\b', str(text))  # \b ensures exact number of digits
    return numbers

# Load dataframe with HSN codes
df = pd.read_pickle('/content/HSNCODE.pkl')

# Parse JSON data
data = json.loads(json_data)

# Iterate over each item
for item in data:
    description = item.get('description', '')  # Placeholder for description retrieval
    # First, try to extract HSN codes from description
    hsn_codes = extract_hsn_codes(description)
    if hsn_codes:
        # If HSN codes are found, use them
        item['HSN_Code'] = hsn_codes[0]  # Assuming only one HSN code per item
    else:
        # If HSN codes are not found in description, search in other fields
        numbers = []
        for key, value in item.items():
            numbers.extend(find_numbers(value))

        valid_hsn = None
        for x in numbers:
            x = int(x)
            is_present = df.isin([x]).any().any()
            if is_present:
                valid_hsn = x

        # If valid HSN code found, add it to the item
        if valid_hsn:
            item['HSN_Code'] = valid_hsn  # Assuming only one HSN code per item
        else:
            item['HSN_Code'] = None

# Print the modified JSON data
print(json.dumps(data, indent=2))
```
